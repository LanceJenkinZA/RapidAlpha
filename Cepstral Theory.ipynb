```python
from BaseDelegate import BaseDelegate
from pylab import *
bd = BaseDelegate()
```

```python
measurement_file = "../testdata/report/reflective/120927/1.db"
alpha = bd.loadAbsorptionCoefficient(measurement_file)
```

```python
t = arange(0, len(alpha.microphone_cepstrum) / (44100.0 / 7.0), 1 / (44100.0 / 7.0))
plot(1000 * t, alpha.microphone_cepstrum, ls="-", c="k", label="Microphone Cepstrum")
plot(1000 * t, alpha.generator_cepstrum, ls="--", c="k", label="Generator Cepstrum")
xlim(0, 12)
ylim(-1, 1)
grid(True)
xlabel("Quefrency (ms)")
ylabel("$c[n]$")
ax = gca()
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_major_locator(MultipleLocator(0.1))
legend(bbox_to_anchor=(0., -.2, 1., .102), loc=3, ncol=3, mode="expand", borderaxespad=0., handlelength=5, prop={'size':8})
tight_layout()
savefig("./mic_gen_cepstrum.eps", bbox_inches='tight')
```
