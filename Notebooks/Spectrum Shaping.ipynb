```python
fs = 8000.0
N = 2 ** 10
M = 10
TS = zeros(N)
k = arange(N)
w = -1 * ones(M)
w[0] = 1
for m in arange(M):
    TS += w[m] * cos(2 * pi * k * m / N)
    
plot(linspace(0, fs / 2, N / 2), TS[:N/2])

ax = gca()
ax.yaxis.set_ticklabels(10 ** arange(-8.0 + 5, 4))

ylabel("$|S(f)|$")
xlabel("Frequency (Hz)")
grid(True)
tight_layout()
savefig("./target_spectrum.eps")

c = ifft(TS)
figure()
plot(1000 * arange(0, len(c) / fs, 1 / fs), c)
xlim(0, 5)
ylabel("c[n]")
xlabel("Time (ms)")
grid(True)
tight_layout()
savefig("./target_cepstrum.eps")
```
