```python
from pylab import *
from scipy.signal import butter, lfilter
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes

# Get signal parameters
f_0 = 0
f_1 = 6400
T = 0.1
sample_rate = 44100.0
fft_size = 2 ** 14
signal_length = 0.1

# Generate time vector
t = arange(0, signal_length, 1 / sample_rate)

# Generate the signal from 0 to Nyquist frequency
a = pi * ((sample_rate / 2) - f_0) / (2 * T)
b = 2 * pi * f_0

s = sin((a * t + b) * t)



# Determine the spectrum
S = fft(s, fft_size)

# Inverse of the magnitude spectrum
iaS = abs(S) ** -1
figure()
plot(fftfreq(fft_size, 1 / sample_rate)[:fft_size / 2], iaS[:fft_size / 2])
ylabel(r"$\left| S\left(\ f \right)\right| ^ {-1}$")
xlabel("Frequency (Hz)")
grid(True)
tight_layout()
savefig("inverse_magnitude_spectrum.eps")

# c, similar to the cepstrum, is the inverse of the logarithmic inverse
# magnitude spectrum
c = ifft(log(iaS))

# Window c to produce m
m = r_[c[0], 2 * c[1:len(S) / 2 - 1], c[len(S) / 2], zeros(len(S) / 2)]

# Determine the spectrum of the windowed 'cepstrum'
M = fft(m, fft_size)

# Determine the minimum phase inverse filter
iSmp = exp(M)

# Determine the minimum phase spectrum
Smp = S * iSmp

# Determin the minimum phase signal
smp = ifft(Smp)

# smp will have fft_size samples, which could be very long
# reduce to length of the signal specified
smp = smp[:len(t)]

# Low pass filter the signal to the upper frequency
[b, a] = butter(8, f_1 / (sample_rate / 2), btype="low")
smp = lfilter(b, a, smp)


Smp = abs(fft(smp, fft_size))
a = pi * ((f_1) - f_0) / (T)
b = 2 * pi * f_0

s = sin((a * t + b) * t)
figure()
plot(1000 * t, s)
xlabel("Time (ms)")
ylabel("Amplitude")
S = fft(s, fft_size)
grid(True)
tight_layout()
savefig("swept_sine.eps")
figure()
semilogx(fftfreq(fft_size, 1 / sample_rate), 20*log10(abs(S) / max(abs(S))), label="Swept Sine")
semilogx(fftfreq(fft_size, 1 / sample_rate), 20*log10(Smp / max(Smp)), ls="--", color="black", label="Low Pass Swept Sine")
ylabel(r"$20 \log\left| S\left(\ f \right)\right|$ (dB)")
xlabel("Frequency (Hz)")
xlim(100, 10000)
ylim(-10, 1)
grid(True)
legend(bbox_to_anchor=(0., -.23, 1., .102), loc=3, ncol=3, mode="expand", borderaxespad=0., handlelength=5, prop={'size':8})
tight_layout()
savefig("modified_spectrum.eps", bbox_inches='tight')

# Normalize so that the maximum value is 1
smp /= max(abs(smp))

figure()
plot(1000 * t, s, color="grey")
plot(1000 * t, smp, color="black")
xlabel("Time (ms)")
ylabel("Amplitude")
tight_layout()
savefig("modified_signal.eps")
```
