# Helmholtz Resonator

## Constants

```python
c = 340 # speed of sound
N = 52 # Number of holes
a = 2.5e-3 # radius of holes
A = 51.5e-3 # radius of sample
t = 3e-3 # Thickness of sample
L = arange(0.001, 100e-3, 0.001) # Range of valid lengths behind sample

S = pi * a ** 2 # Surface area of hole
V = (pi * A ** 2 * L) / N # Effective volume of resonator
l = t + 1.7 * a # Effective length of neck
```

## Code

```python
from pylab import *
```

```python
f_0 = (c / (2 * pi) ) * sqrt(S / (V * l))
plot(1000 * L, f_0)
grid(True)
title("Resonant Frequency with Varying Backing Length")
ylabel("Resonant Frequency (Hz)")
xlabel("Backing Length (mm)")
```
