# Low Pass Swept Sine

```python
quiet_file = "../testdata/report/asphalt-quiet-lpss/121010/8.db"
noisy_file = "../testdata/report/asphalt-noisy-lpss/121010/8.db"

quiet_alpha = bd.loadAbsorptionCoefficient(quiet_file)
noisy_alpha = bd.loadAbsorptionCoefficient(noisy_file)

semilogx(fftfreq(len(quiet_alpha.alpha), 1 / (44100.0 / 7.0)), quiet_alpha.alpha)
semilogx(fftfreq(len(noisy_alpha.alpha), 1 / (44100.0 / 7.0)), noisy_alpha.alpha, color="k", ls="--")

axes = gca()
axes.set_xlim([100, 2100])
axes.set_xticks([100, 125, 160, 200, 250, 315, 400, 500, 630, 800, 1000, 1250, 1600, 2000])
axes.set_xticklabels([100, 125, 160, 200, 250, 315, 400, 500, 630, 800, 1000, 1250, 1600, 2000])
axes.xaxis.set_minor_locator(NullLocator())
axes.set_ylim([0, 1])
axes.yaxis.set_major_locator(MultipleLocator(0.1))
axes.grid(color="grey", linestyle="--")
axes.set_xlabel("Frequency (Hz)")
axes.set_ylabel(r"Absorption Coefficient")
fig = gcf()
fig.subplots_adjust(bottom=0.1, top=0.98, right=0.98, left=0.05)
N = len(noisy_alpha.alpha)
n_200 = noisy_alpha.alpha[200 * (7.0 * N ) / (44100.0)]
q_200 = quiet_alpha.alpha[200 * (7.0 * N ) / (44100.0)]
diff = abs(n_200 - q_200)
annotate("%.2f"%(diff), xy=(200, n_200), xycoords="data", xytext=(200, n_200 + 0.08), ha="center", va="bottom",
            arrowprops=dict(arrowstyle="->"))
annotate(" ", xy=(200, q_200), xycoords="data", xytext=(200, q_200 - 0.08), ha="center", va="bottom",
            arrowprops=dict(arrowstyle="->"))

n_250 = noisy_alpha.alpha[250 * (7.0 * N ) / (44100.0)]
q_250 = quiet_alpha.alpha[250 * (7.0 * N ) / (44100.0)]
diff = abs(n_250 - q_250)
annotate("%.2f"%(diff), xy=(250, n_250), xycoords="data", xytext=(250, n_250 + 0.08), ha="center", va="bottom",
            arrowprops=dict(arrowstyle="->"))
annotate(" ", xy=(250, q_250), xycoords="data", xytext=(250, q_250 - 0.08), ha="center", va="bottom",
            arrowprops=dict(arrowstyle="->"))
legend(bbox_to_anchor=(0., -.2, 1., .102), loc=3, ncol=3, mode="expand", borderaxespad=0., handlelength=5, prop={'size':8})
tight_layout()
#savefig("noise_imunity_lpss.eps", bbox_inches='tight')
figure()
```

# Inverse Repeat Sequence

```python
quiet_file = "../testdata/report/asphalt-quiet-irs-1/121010/3.db"
noisy_file = "../testdata/report/asphalt-noisy-irs/121010/2.db"

quiet_alpha = bd.loadAbsorptionCoefficient(quiet_file)
noisy_alpha = bd.loadAbsorptionCoefficient(noisy_file)

semilogx(fftfreq(len(quiet_alpha.alpha), 1 / (44100.0 / 7.0)), quiet_alpha.alpha, label="Quiet Environment")
semilogx(fftfreq(len(noisy_alpha.alpha), 1 / (44100.0 / 7.0)), noisy_alpha.alpha, color="k", ls="--", label="Noisy Environment")

axes = gca()
axes.set_xlim([100, 2100])
axes.set_xticks([100, 125, 160, 200, 250, 315, 400, 500, 630, 800, 1000, 1250, 1600, 2000])
axes.set_xticklabels([100, 125, 160, 200, 250, 315, 400, 500, 630, 800, 1000, 1250, 1600, 2000])
axes.xaxis.set_minor_locator(NullLocator())
axes.set_ylim([0, 1])
axes.yaxis.set_major_locator(MultipleLocator(0.1))
axes.grid(color="grey", linestyle="--")
axes.set_xlabel("Frequency (Hz)")
axes.set_ylabel(r"Absorption Coefficient")

N = len(noisy_alpha.alpha)

# 200 Hz
n = noisy_alpha.alpha[200 * (7.0 * N ) / (44100.0)]
q = quiet_alpha.alpha[200 * (7.0 * N ) / (44100.0)]
diff = abs(n - q)
annotate("%.2f"%(diff), xy=(200, max(n, q)), xycoords="data", xytext=(200, max(n, q) + 0.08), ha="center", va="bottom",
            arrowprops=dict(arrowstyle="->"))
annotate(" ", xy=(200, min(n, q)), xycoords="data", xytext=(200, min(n, q) - 0.08), ha="center", va="bottom",
            arrowprops=dict(arrowstyle="->"))

# 1000 Hz
n = noisy_alpha.alpha[1000 * (7.0 * N ) / (44100.0)]
q = quiet_alpha.alpha[1000 * (7.0 * N ) / (44100.0)]
diff = abs(n - q)
annotate("%.2f"%(diff), xy=(1000, max(n, q)), xycoords="data", xytext=(1000, max(n, q) + 0.08), ha="center", va="bottom",
            arrowprops=dict(arrowstyle="->"))
annotate(" ", xy=(1000, min(n, q)), xycoords="data", xytext=(1000, min(n, q) - 0.08), ha="center", va="bottom",
            arrowprops=dict(arrowstyle="->"))

# 1600 Hz
f_point = 1600
n = noisy_alpha.alpha[f_point * (7.0 * N ) / (44100.0)]
q = quiet_alpha.alpha[f_point * (7.0 * N ) / (44100.0)]
diff = abs(n - q)
annotate(" ", xy=(f_point, max(n, q)), xycoords="data", xytext=(f_point, max(n, q) + 0.08), ha="center", va="bottom",
            arrowprops=dict(arrowstyle="->"))
annotate("%.2f"%(diff), xy=(f_point, min(n, q)), xycoords="data", xytext=(f_point, min(n, q) - 0.08), ha="center", va="top",
            arrowprops=dict(arrowstyle="->"))
legend(bbox_to_anchor=(0., -.2, 1., .102), loc=3, ncol=3, mode="expand", borderaxespad=0., handlelength=5, prop={'size':8})
tight_layout()
#savefig("noise_imunity_irs.eps", bbox_inches='tight')
```
