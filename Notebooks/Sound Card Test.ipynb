```python
octaves = 10.0
tones = 10.0
f_base = 16.0
fs = 44100.0

S = zeros(fs)
for octave in arange(octaves):
    for tone_index in arange(tones):
        # Determine current tone
        tone = round(octave * tones + tone_index) / tones
        
        #tone=round((fslide+ j)*NTONE+(double)jj)/(double)NTONE;
        #k=((int)((pow(2.,tone) * FBASE)));
        # and the index of the tone
        k = int(f_base * 2 ** tone)
        
        if k > fs / 2.0:
            print "k: %s octave: %s tone_index: %s tone: %s" %(k, octave, tone_index, tone)
        S[k] = sin(2 * pi * 1000 * k * tone_index / fs) + 1j * cos(2 * pi * 1000 * k * tone_index / fs)
        # The signal is real
        S[fs - k] = conj(S[k])
s = fft(S)
s /= max(s)
plot(s)


```

```python
def thd(x, freq, NFFT, fs, order):
    # determine maximum order
    order = min(int(order), floor((float(fs) / 2) / float(freq)))

    # frequency resolution of the PSD
    df = fs / float(NFFT)
    
    # bin of the frequency
    bin = freq / df
    
    # sum up harmonics
    P_harmonics = 0
    for harmonic in arange(2, order + 1):
        P_harmonics += sum(x[bin * harmonic - 1:bin * harmonic + 2])
    
    return 100 * sqrt(P_harmonics / sum(x[bin - 1:bin + 2]))
```
