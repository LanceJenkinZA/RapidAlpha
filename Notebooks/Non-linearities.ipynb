```python
def thd(x, freq, NFFT, fs, order):
    # determine maximum order
    order = min(int(order), floor((float(fs) / 2) / float(freq)))

    # frequency resolution of the PSD
    df = fs / float(NFFT)
    
    # bin of the frequency
    bin = freq / df
    
    # sum up harmonics
    P_harmonics = 0
    for harmonic in arange(2, order + 1):
        P_harmonics += sum(x[bin * harmonic - 1:bin * harmonic + 2])
    
    return 100 * sqrt(P_harmonics / sum(x[bin - 1:bin + 2]))
```

```python
from matplotlib.mlab import psd

A = 0.1
fs = 44100.0
T = 2
T1 = 0.75
T2 = 1.75
t = arange(0, 2, 1 / fs)
THD_mic = []
THD_gen = []
NFFT = 4410
f_range = arange(50, 5050, 50)
for freq in f_range:
    left = loadtxt("../testdata/report/non-linearities/121022-left-%s.txt" % (freq))
    right = loadtxt("../testdata/report/non-linearities/121022-right-%s.txt" % (freq))
    l = left[int(T1 * fs): int(T2 * fs)]
    r = right[int(T1 * fs): int(T2 * fs)]
    
    Pxx, f = psd(l, NFFT=NFFT, Fs=fs)
    
    THD_mic.append(thd(Pxx, freq, NFFT, fs, 5))
    Pxx, f = psd(r, NFFT=NFFT, Fs=fs)
    THD_gen.append(thd(Pxx, freq, NFFT, fs, 5))
```

```python
semilogx(f_range, THD_mic)
ylim(0, 1)
xlim(100, 2000)
grid(True)
ylabel("THD %")
xlabel("Frequency (Hz)")
axes = gca()
axes.yaxis.set_major_locator(MultipleLocator(0.1))

axes.axes.set_xticks([100, 125, 160, 200, 250, 315, 400, 500, 
                        630, 800, 1000, 1250, 1600, 2000])
axes.set_xticklabels([100, 125, 160, 200, 250, 315, 400, 500, 
                        630, 800, 1000, 1250, 1600, 2000], rotation=30)

tight_layout()
```

```python
from MlsDb import MlsDb
from scipy.signal import butter, lfilter, decimate

q = 7

mlsdb = MlsDb()
mls = mlsdb.getMls(14)
mls = 2 * mls -1 
[b, a] = butter(1, 200 / (44100.0 / 2.0), btype="high")
x = lfilter(b, a, mls)

y = mlsdb.getSystemResponse(x, 14)
y = decimate(y, q)
t = arange(0, len(y) * q / 44100.0, q / 44100.0 )

c = ifft(log(abs(fft(y))))
plot (1000 * t, c, label="No Non-Linear Distortion")

x2 = x + 0.001 * x ** 2
y2 = mlsdb.getSystemResponse(x2, 14)
y2 = decimate(y2, q)
c2 = ifft(log(abs(fft(y2))))
plot (1000 * t, c2, ls="--", color="k", label="0.1% 2nd Order Distortion")

x3 = x + 0.01 * x ** 2
y3 = mlsdb.getSystemResponse(x3, 14)
y3 = decimate(y3, q)
c3 = ifft(log(abs(fft(y3))))
plot (1000 * t, c3, ls="-.", color="k", label="1% 2nd Order Distortion")

xlim(0, 12)
ylim(-0.11, 0.011)
legend(bbox_to_anchor=(0., -.23, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0., handlelength=5, prop={'size':8})
grid(True)

bin_loc = len(y) * 1000 * 8 / 44100.0 / 2.0
ax = gca()
bin_loc = 44100.0 * 8 / 1000.0 / q
print abs(c2[bin_loc] - c[bin_loc])
#ax.text(9, -0.004, "28 samples", ha="center", va="bottom", size=10)
ax.annotate("", xy=(8, c2[bin_loc]), xycoords="data", ha="center", va="center",xytext=(8, 0.002),
            arrowprops=dict(arrowstyle="->"))
ax.annotate("%.3f" % (abs(c2[bin_loc] - c[bin_loc])), xy=(8, c[bin_loc]), xycoords="data", ha="center", va="center", xytext=(8, c[bin_loc] - 0.01),
            arrowprops=dict(arrowstyle="->"))
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_major_locator(MultipleLocator(0.01))
xlabel("Quefrency (ms)")
ylabel("c[n]")
tight_layout()
savefig("offset_cepstrum.eps", bbox_inches='tight')
```
