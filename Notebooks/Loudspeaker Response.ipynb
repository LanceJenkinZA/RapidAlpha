```python
from pylab import *
from scipy.signal import lfilter, butter
from BaseDelegate import BaseDelegate
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
```

```python
# sample rate
fs = 44100.0

# time
t = arange(0, 10, 1 / fs)

# Excitation signal
x = r_[1, zeros(len(t) - 1)]
cepstrum = lambda _: ifft(log(abs(fft(_))))

# Loudspeaker HPF
[ls_b, ls_a] = butter(1, 2001 / (fs / 2.0), btype="high")
x_ls = lfilter(ls_b, ls_a, x.copy())
y = x_ls

plot(1000 * t, cepstrum(y), ls="-", c="k", label="2000 Hz cut off")

[ls_b, ls_a] = butter(1, 200 / (fs / 2.0), btype="high")
x_ls = lfilter(ls_b, ls_a, x.copy())
y = x_ls

plot(1000 * t, cepstrum(y), ls="--", c="k", label="200 Hz cut off")

[ls_b, ls_a] = butter(1, 20 / (fs / 2.0), btype="high")
x_ls = lfilter(ls_b, ls_a, x.copy())
y = x_ls

plot(1000 * t, cepstrum(y), ls="-.", c="k", label="20 Hz cut off")
xlim(0, 12)
ylim(-0.02, 0.006)
legend()
grid(True)
xlabel("Quefrency (ms)")
ylabel("$c[n]$")

ax = gca()
ax.xaxis.set_major_locator(MultipleLocator(1))
legend(bbox_to_anchor=(0., -.2, 1., .102), loc=3, ncol=3, mode="expand", borderaxespad=0., handlelength=5, prop={'size':8})
tight_layout()

savefig("loudspeaker_cepstrum_hpf.eps", bbox_inches='tight')
```

```python
bd = BaseDelegate()
measurement_file = "../testdata/speaker_response/120906_no_modifications.fdb"
freq = bd.loadFrequencyResponse(measurement_file)
sample_rate = 44100.0
frequency_axis = fftfreq(len(freq.frequency_response), 1 / sample_rate)
SPL = 10 * log10(abs(freq.frequency_response))
SPL -= max(SPL)
semilogx(frequency_axis, SPL)
# Annotate the graph
xlabel("Frequency (Hz)")
ylabel(r"SPL (dB)")

# Format the graph
xlim(10, 10000)
ylim(-30, 0)
axes = gca()
grid(True)

tight_layout()
savefig("./loudspeaker_response_no_modifications.eps")

figure()
plot(freq.impulse_response)
```

```python
# sample rate
fs = 44100.0

# time
t = arange(0, 10, 1 / fs)

# Excitation signal
x = r_[1, zeros(len(t) - 1)]
cepstrum = lambda _: ifft(log(abs(fft(_))))

# Loudspeaker HPF
[ls_b, ls_a] = butter(1, 200 / (fs / 2.0), btype="high")
x_ls = lfilter(ls_b, ls_a, x.copy())
y = x_ls

plot(1000 * t, cepstrum(y), ls="-", c="k", label="1st Order")

[ls_b, ls_a] = butter(2, 200 / (fs / 2.0), btype="high")
x_ls = lfilter(ls_b, ls_a, x.copy())
y = x_ls

plot(1000 * t, cepstrum(y), ls="--", c="k", label="2nd Order")

[ls_b, ls_a] = butter(3, 200 / (fs / 2.0), btype="high")
x_ls = lfilter(ls_b, ls_a, x.copy())
y = x_ls

plot(1000 * t, cepstrum(y), ls="-.", c="k", label="3rd Order")
xlim(0, 12)
ylim(-0.02, 0.02)
legend()
grid(True)
xlabel("Quefrency (ms)")
ylabel("$c[n]$")
ax = gca()
ax.xaxis.set_major_locator(MultipleLocator(1))
savefig("loudspeaker_cepstrum_varying_order.eps")
```

```python
h = freq.impulse_response
h /= max(abs(h))

t = arange(0, len(h) / fs, 1 / fs)
plot(1000 * t, h)

grid(True)
ax = gca()
xlabel("Time (ms)")
ylabel("h[n]")

axins = inset_axes(ax, 4, 1, loc=1) # zoom = 6
axins.plot(1000 * t, h)
axins.grid(True)
axins.yaxis.set_ticklabels([""])
axins.xaxis.set_ticklabels([""])
axins.set_xlim(20, 22.45)
axins.set_ylim(-0.05, 0.05)
mark_inset(ax, axins, loc1=3, loc2=4, fc="none", ec="0.5")

ax.text(17, 0.45, "10 cycles in 2.45 ms = 4096 Hz", ha="center", va="top")
tight_layout()
savefig("loudspeaker_impulse_response.eps")
```
