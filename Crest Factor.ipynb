```python
from BaseDelegate import BaseDelegate
import copy
bd = BaseDelegate()
```

```python
lpss_file = "../testdata/report/asphalt-quiet-lpss/121010/1.db"
irs_file = "../testdata/report/asphalt-quiet-irs/121010/1.db"

lpss = bd.loadAbsorptionCoefficient(lpss_file)
irs = bd.loadAbsorptionCoefficient(irs_file)

lpss_mic = copy.deepcopy(lpss.microphone_signals[1])
irs_mic = copy.deepcopy(irs.microphone_signals[1])
```

```python
fs = 44100.0

t = arange(0, len(lpss_mic) / fs, 1 / fs)
plot(1000 * t, lpss_mic)
xlim(300, 700)
grid(True)
ylabel("Amplitude")
xlabel("Time (ms)")
tight_layout()

savefig("emitted_low_pass_swept_sine.eps")

figure()
t = arange(0, len(irs_mic) / fs, 1 / fs)
plot(1000 * t, irs_mic)
xlim(300, 700)
grid(True)
ylabel("Amplitude")
xlabel("Time (ms)")
tight_layout()

savefig("emitted_inverse_repeat_sequence.eps")
```

```python
rms = lambda x: sqrt((1 / float(len(x))) * sum(x ** 2))
crest_factor = lambda x: max(x) / rms(x)
```

```python
print crest_factor(lpss_mic[0.34 * fs: 0.46 * fs])
print crest_factor(irs_mic[0.3 * fs: 5 * fs])
```
