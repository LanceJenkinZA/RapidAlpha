```python
from BaseDelegate import BaseDelegate
bd = BaseDelegate()
```

```python
measurement_file = "../testdata/report/reflective/120905/1.db"
alpha = bd.loadAbsorptionCoefficient(measurement_file)
```

```python
plot(alpha.power_cepstrum)
xlim(0, 170)
annotate("Direct Cepstra", xy=(10,-0.1), xycoords='data', xytext=(10,-0.5), arrowprops=dict(arrowstyle="->"))
annotate("Sample Impulse Response", xy=(37,0.8), xycoords='data', xytext=(30,1.5), arrowprops=dict(arrowstyle="->"))
annotate("Cone Reflection", xy=(74,0.2), xycoords='data', xytext=(74,0.8), arrowprops=dict(arrowstyle="->"))
annotate("1st Rahmonic", xy=(72,-0.3), xycoords='data', xytext=(73,-0.8), arrowprops=dict(arrowstyle="->"), ha="center")
annotate("2nd Rahmonic", xy=(107,-0.05), xycoords='data', xytext=(108,-0.5), arrowprops=dict(arrowstyle="->"), ha="center")
annotate("3rd Rahmonic", xy=(143,-0.2), xycoords='data', xytext=(144,-0.8), arrowprops=dict(arrowstyle="->"), ha="center")
grid(True)
xlabel("Quefrency (samples)")
ylabel("Amplitude")
tight_layout()
savefig("./annotated_cepstrum.eps")
```

```python
plot(alpha.power_cepstrum)
xlim(0, 100)
offset = 33
win_t = offset + arange(0, len(alpha.window))
plot(r_[offset, win_t], r_[0, alpha.window], ls="--", color="k", lw=2)
grid(True)
xlabel("Quefrency (samples)")
ylabel("Amplitude")
tight_layout()
savefig("./windowed_cepstrum.eps")
```
